```python
from pymongo import MongoClient

# Cadena de conexión a MongoDB Atlas
mongo_uri = "mongodb+srv://proyecto:proyecto@cluster0.fwvbnih.mongodb.net/?retryWrites=true&w=majority"

# Conectar a MongoDB Atlas
client = MongoClient(mongo_uri)

# Obtener la lista de nombres de todas las colecciones en la base de datos
database_name = "PROYECTO"
database = client[database_name]
collection_names = database.list_collection_names()

# Iterar sobre las colecciones y obtener sus atributos
for collection_name in collection_names:
    collection = database[collection_name]
    print(f"Nombre de la colección: {collection_name}")
    print("Atributos:")
    for document in collection.find().limit(1):
        for key in document.keys():
            print(key)

    print("-----")
```

```python
import pandas as pd

# Obtener las fechas únicas de 2022 y 2023
fechas_2022_2023 = df1['Date'].unique()

# Generar las fechas de 2021
fechas_2021 = pd.date_range(start='2021-01-01', end='2021-12-31')

# Calcular el promedio de las columnas para las fechas de 2022 y 2023
promedio_por_columna = df1.groupby('Date').mean()

# Crear un DataFrame con las fechas de 2021 y los promedios por columna
df_2021 = pd.DataFrame({
    'Date': fechas_2021,
    'Viajes por dia': promedio_por_columna['Viajes por dia'].mean(),
    'Pasajeros por dia': promedio_por_columna['Pasajeros por dia'].mean(),
    'Distancia(millas)': promedio_por_columna['Distancia(millas)'].mean(),
    'Tarifario por dia($USD)': promedio_por_columna['Tarifario por dia($USD)'].mean(),
    'travel_tday_bgh': promedio_por_columna['travel_tday_bgh'].mean(),
    'borough': '',  # Valor sintético para borough
    'Total recaudado por dia($USD)': promedio_por_columna['Total recaudado por dia($USD)'].mean(),
    'Pago con tarjeta': promedio_por_columna['Pago con tarjeta'].mean(),
    'Pago con efectivo': promedio_por_columna['Pago con efectivo'].mean(),
    'Tipo de taxi': ''  # Valor sintético para Tipo de taxi
})

# Concatenar los DataFrames df_2021 y df1
df_final = pd.concat([df_2021, df1], ignore_index=True)

# Guardar el DataFrame resultante en un nuevo archivo CSV
df_final.to_csv('nuevo_dataframe.csv', index=False)
```
